```python
import pandas as pd
import numpy as np

# 创建示例数据（假设数据已经加载到DataFrame df 中）
# 请将示例数据替换为你的实际数据
data = {
    'id': [236] * 24,
    'time': pd.to_datetime([
        '2015-05-22 14:40:56', '2015-05-22 14:42:23', '2015-05-22 14:44:31',
        '2015-05-22 14:49:31', '2015-05-22 14:54:31', '2015-05-22 14:59:31',
        '2015-05-22 15:04:31', '2015-05-22 15:09:31', '2015-05-22 15:14:31',
        '2015-05-22 15:19:31', '2015-05-22 15:24:31', '2015-05-22 15:29:31',
        '2015-05-22 15:34:31', '2015-05-22 15:39:31', '2015-05-22 15:44:31',
        '2015-05-22 15:49:31', '2015-05-22 15:54:31', '2015-05-22 15:59:31',
        '2015-05-22 16:04:31', '2015-05-22 16:09:31', '2015-05-22 16:14:31',
        '2015-05-22 16:19:31', '2015-05-22 16:24:31', '2015-05-22 16:29:31'
    ]),
    'gl': [183.0, 203.0, 194.0, 191.0, 199.0, 207.0, 211.0, 216.0, 217.0, 213.0, 210.0,
           208.0, 206.0, 203.0, 200.0, 198.0, 197.0, 200.0, 201.0, 207.0, 208.0, 212.0,
           206.0, 204.0]
}
df = pd.DataFrame(data)

# 将时间列转换为日期时间类型
df['time'] = pd.to_datetime(df['time'])

# 获取每个时间的小时、分钟和秒
df['hour'] = df['time'].dt.hour

df['minute'] = df['time'].dt.minute

print(df)
# 创建用于插值的时间序列
complete_time_range = pd.date_range(start=df['time'].min(), end=df['time'].max(), freq='5T')

# 对每个小时的数据进行插值
result = []
for hour in df['hour'].unique():
    hour_data = df[df['hour'] == hour]
    
    if len(hour_data) < 12:
        # 进行插值
        interpolated_time_range = complete_time_range[(complete_time_range.hour == hour) & (complete_time_range.minute % 5 == 0)]
        interpolated_gl = np.interp(
            interpolated_time_range.minute, hour_data['minute'], hour_data['gl'], left=np.nan, right=np.nan
        )
        
        interpolated_data = pd.DataFrame({
            'time': interpolated_time_range,
            'id': hour_data['id'].iloc[0],
            'gl': interpolated_gl
        })
        
        result.append(interpolated_data)
    else:
        result.append(hour_data)

# 合并插值后的数据
final_result = pd.concat(result)

print(final_result)
```
